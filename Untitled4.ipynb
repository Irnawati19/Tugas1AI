```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#mebuat data dari file csv
df = pd.read_csv('D:/irnawati/heart.csv')
```

```python
#menambahkan beberapa baris pertama
print(df.head())
```

```python
#mengghitung jumlah nilai yang hilang
print(df.isnull().sum())
```

```python
#menghapus baris yang  mengandung  nilai yang hilang
df.dropna(inplace=True)
```

```python
# Definisikan fitur dan target
x = df.drop('output',axis=1)
y = df[['output']]
```

```python
# Membagi data menjadi train dan test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Membuat model Decision Tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='gini',random_state=42,max_depth=3,min_samples_leaf=5)
```

```python
# Melatih model dengan data pelatihan
model.fit(x_train,y_train)
```

```python
# Membuat model Decision Tree
model = DecisionTreeClassifier(random_state=42)
```

```python
# Melatih model dengan data pelatihan
model.fit(x_train, y_train)
```

```python
# Membuat prediksi dengan data uji
y_pred = model.predict(x_test)
```

```python
# Prediksi label untuk data tes
y_pred = model.predict(x_test)
y_pred
```

```python
# Impor fungsi classification_report dari modul sklearn.metrics
from sklearn.metrics import classification_report
```

```python
# Impor fungsi confusion_matrix dari modul sklearn.metrics
from sklearn.metrics import confusion_matrix
```

```python
# Evaluasi model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(classification_report(y_test, y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```
